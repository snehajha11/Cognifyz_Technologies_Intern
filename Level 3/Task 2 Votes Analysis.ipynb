Task - 2 : Votes Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset .csv")

df.head(3)
```

1. Identify the restaurants with the highest and lowest number of votes.

```python
df['Votes'].describe()
```

```python
high_no_votes = df.nlargest(1, 'Votes')
```

```python
print("Restaurant with the highest number of votes:")
print(high_no_votes[['Restaurant Name', 'Votes']])
```

```python
low_no_votes = df[df['Votes'] == df['Votes'].min()]

print("Restaurants with the lowest number of votes:")
print(low_no_votes[['Restaurant Name', 'Votes']])
```

2. Analyze if there is a correlation between the number of votes and the rating of arestaurant

```python
votes = df['Votes']
rating = df['Aggregate rating']

correlation_coefficient = votes.corr(rating)

plt.figure(figsize=(8, 4))
sns.scatterplot(x=votes, y=rating, alpha=0.5)
plt.title(f'Correlation between Votes and Rating : {correlation_coefficient:.2f}')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.show()
```
